```python
import gym
import numpy as np
import random
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#objective is to have an agent learn to navigate to the goal without moving onto a hole. 
#up, down, left, or right
#wind which occasionally blows the agent onto a space they didnâ€™t choose. 
#reward at every step is 0, except for entering the goal, which provides a reward of 1
env = gym.make('FrozenLake-v0')
```

```python
#clears the default graph stack and resets the global default graph.
tf.reset_default_graph()
```

```python
#These lines establish the feed-forward part of the network used to choose actions
#inserts a placeholder for a tensor that will be always fed.
#shape: The shape of the tensor to be fed
#Multiplies matrix a by matrix b, producing a * b.
#returns the index with the largest value across axes of a tensor. (deprecated arguments)
inputs1 = tf.placeholder(shape=[1,16],dtype=tf.float32)
W = tf.Variable(tf.random_uniform([16,4],0,0.01))
Qout = tf.matmul(inputs1,W)
predict = tf.argmax(Qout,1)

#Below we obtain the loss by taking the sum of squares difference between the target and prediction Q values.
#computes the sum of elements across dimensions of a tensor. (deprecated arguments)
#optimizer that implements the gradient descent algorithm.
nextQ = tf.placeholder(shape=[1,4],dtype=tf.float32)
loss = tf.reduce_sum(tf.square(nextQ - Qout))
trainer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
updateModel = trainer.minimize(loss)
```

```python
#an Op that initializes global variables in the graph.
init = tf.global_variables_initializer()

# Set learning parameters
discount_rate = .99
epsilon = 0.1
num_episodes = 10
#create lists to contain total rewards and steps per episode
tsList = []
rewardTable = []
with tf.Session() as sess:
    sess.run(init)
    for i in range(num_episodes):
        #Reset environment and get first new observation
        state = env.reset()
        totalReward = 0
        done = False
        ts = 0
        #The Q-Network
        while ts < 99:
            env.render()
            ts+=1
            #Choose an action by greedily (with e chance of random action) from the Q-network
            action,allQ = sess.run([predict,Qout],feed_dict={inputs1:np.identity(16)[state:state+1]})
            if np.random.rand(1) < epsilon:
                action[0] = env.action_space.sample()
            #Get new state and reward from environment
            next_state,reward,done,_ = env.step(action[0])
            #Obtain the Q' values by feeding the new state through our network
            Q1 = sess.run(Qout,feed_dict={inputs1:np.identity(16)[next_state:next_state+1]})
            #Obtain maxQ' and set our target value for chosen action.
            maxQ1 = np.max(Q1)
            targetQ = allQ
            targetQ[0,action[0]] = reward + discount_rate*maxQ1
            #Train our network using target and predicted Q values
            _,W1 = sess.run([updateModel,W],feed_dict={inputs1:np.identity(16)[state:state+1],nextQ:targetQ})
            totalReward += reward
            state = next_state
            if done == True:
                #Reduce chance of random action as we train the model.
                epsilon = 1./((i/50) + 10)
                break
        tsList.append(ts)
        rewardTable.append(totalReward)
        # end of game
        print('game over')
        env.close()
print "Percent of succesful episodes: " + str(sum(rewardTable)/num_episodes) + "%"
```
